```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
```

```python
player_list = pd.read_csv('NBA Season Data (Original).csv')
```

```python
player_list_by_player = player_list.set_index('Player').drop('0')
player_list_by_player
```

```python
players_2015 = player_list.set_index('Year').loc[2015]
players_2015.columns.values
```

```python
players_2015_indexed = players_2015.set_index(['Tm', 'Player','PER'])
```

```python
## Calculate the top 10 players in the league by PER (we exclude those who played too few games).
players_2015_top_per = players_2015_indexed.sort_index(level='PER', ascending=False)
players_2015_top_per = players_2015_top_per[players_2015_top_per.G>50]
players_2015_top_per.head(10)
```
